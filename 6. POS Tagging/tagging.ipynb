## **Understanding Part-of-Speech Tagging**

Part-of-Speech Tagging involves labeling each word in a sentence with its corresponding part of speech, such as noun, verb, adjective, etc. This helps in understanding the grammatical structure and meaning of the sentence.

### NLTK uses the Penn Treebank POS tags, which include:

- NN - Noun, singular or mass
- NNS - Noun, plural
- NNP - Proper noun, singular
- NNPS - Proper noun, plural
- VB - Verb, base form
- VBD - Verb, past tense
- VBG - Verb, gerund or present participle
- VBN - Verb, past participle
- VBP - Verb, non-3rd person singular present
- VBZ - Verb, 3rd person singular present
- JJ - Adjective
- RB - Adverb
- IN - Preposition or subordinating conjunction
- DT - Determiner
- PRP - Personal pronoun

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
text = "The quick brown fox jumps over the lazy dog."

#tokenising this first

tokens = word_tokenize(text)
print("The tokens are as follows: \n",tokens)

tags = nltk.pos_tag(tokens)
print("The tags are as follows: \n",tags)
```

```python
# visualizing the tagging

import matplotlib.pyplot as plt
from collections import Counter

# Count the POS tags
pos_counts = Counter(tag for word, tag in tags)

# Plot the POS tags
plt.figure(figsize=(10, 6))
plt.bar(pos_counts.keys(), pos_counts.values())
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.title('Part-of-Speech Tag Frequency')
plt.show()
```

```python
# tagging for larger paragraph

paragraph = "The quick brown fox jumps over the lazy dog. A fast and agile fox leaped over a sleepy dog."
sentences = nltk.sent_tokenize(paragraph)
for sentence in sentences:
    tokens = word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    print(pos_tags)
```
